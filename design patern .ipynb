```python

        
class CPU:
    def freeze(self):
        print("CPU freezing...")

    def jump(self, position):
        print(f"CPU jumping to position {position}")

    def execute(self):
        print("CPU executing...")

class Memory:
    def load(self, position, data):
        print(f"Memory loading data '{data}' at position {position}")

class HardDrive:
    def read(self, lba, size):
        return f"Data from sector {lba} with size {size}"

class Computer:
    def __init__(self):
        self.cpu = CPU()
        self.memory = Memory()
        self.hard_drive = HardDrive()

    def start(self):
        print("Starting computer...")
        self.cpu.freeze()
        self.memory.load(0, self.hard_drive.read(0, 1024))
        self.cpu.jump(0)
        self.cpu.execute()

    def shutdown(self):
        print("Shutting down computer...")
        self.cpu.freeze()
        print("Computer is off")


```

```python
class CharacterStyle:
    def __init__(self, font, size, color):
        self.font = font
        self.size = size
        self.color = color

class StyleFactory:
    def __init__(self):
        self.styles = {}

    def get_style(self, font, size, color):
        key = (font, size, color)
        if key not in self.styles:
            self.styles[key] = CharacterStyle(font, size, color)
        return self.styles[key]
    
class Character:
    def __init__(self, char, style):
        self.char = char
        self.style = style

    def display(self):
        print(f"Character: '{self.char}', Font: {self.style.font}, Size: {self.style.size}, Color: {self.style.color}")
if __name__ == "__main__":
    # Create the StyleFactory
    style_factory = StyleFactory()

    # Define some styles
    style1 = style_factory.get_style("Arial", 12, "Black")
    style2 = style_factory.get_style("Times New Roman", 14, "Blue")
    style3 = style_factory.get_style("Arial", 12, "Black")  # This should reuse the existing Arial, 12, Black style

    # Create characters with shared styles
    char1 = Character('H', style1)
    char2 = Character('e', style1)
    char3 = Character('l', style2)
    char4 = Character('l', style2)
    char5 = Character('o', style3)

    # Display characters
    char1.display()
    char2.display()
    char3.display()
    char4.display()
    char5.display()

    # Check the number of unique styles created
    print(f"Unique styles created: {len(style_factory.styles)}")

```

```python
class TreeType:
    def __init__(self, name, color):
        self.name = name
        self.color = color

class TreeFactory:
    def __init__(self):
        self._tree_types = {}

    def get_tree_type(self, name, color):
        key = (name, color)
        if key not in self._tree_types:
            self._tree_types[key] = TreeType(name, color)
        return self._tree_types[key]

class Tree:
    def __init__(self, x, y, tree_type):
        self.x = x
        self.y = y
        self.tree_type = tree_type

    def display(self):
        print(f"Tree of type {self.tree_type.name}, color {self.tree_type.color} at ({self.x}, {self.y})")

class Forest:
    def __init__(self):
        self.trees = []
        self.tree_factory = TreeFactory()

    def plant_tree(self, x, y, name, color):
        tree_type = self.tree_factory.get_tree_type(name, color)
        tree = Tree(x, y, tree_type)
        self.trees.append(tree)

    def display_trees(self):
        for tree in self.trees:
            tree.display()

if __name__ == "__main__":
    forest = Forest()

    # Planting trees with shared types
    forest.plant_tree(1, 2, "Oak", "Green")
    forest.plant_tree(3, 4, "Pine", "Green")
    forest.plant_tree(5, 6, "Oak", "Green")
    forest.plant_tree(7, 8, "Cherry", "Pink")
    forest.plant_tree(9, 10, "Pine", "Green")

    forest.display_trees()
```

#### memento 
* use to save internal state of object 
* can restore it without violate
<br><br>
* Memento ==> object save oriiginal
* Caretaker ==> keep tracks , can restore 


```python
class Original:
    def __init__(self, name):
        self.name = name
    def get(self):
        return self.name

class Change:
    def __init__(self, obj):
        self.obj = obj
    def edit(self):
        self.name = "ahmed"  # Temporary assignment
        return self.name
    def save(self):
        self.obj.name = self.edit()  # Update name of the Original instance
    def restore(self):
        self.name = self.obj.get()  # Restore name from the Original instance

a = Original("mohamed")
b = Change(a)

# b.save() 
b.restore()

print(a.name)  #cheack
```

```python
class EditorMemento:
    def __init__(self, content):
        self._content = content

    def get_content(self):
        return self._content

class Editor:
    def __init__(self):
        self._content = ""

    def type(self, text):
        self._content += text

    def delete(self, num_chars):
        self._content = self._content[:-num_chars]

    def get_content(self):
        return self._content

    def create_memento(self):
        return EditorMemento(self._content)

    def restore(self, memento):
        self._content = memento.get_content()

class History:
    def __init__(self):
        self._states = []

    def push(self, state):
        self._states.append(state)

    def pop(self):
        return self._states.pop()

if __name__ == "__main__":
    editor = Editor()
    history = History()

    editor.type("Hello, ")
    history.push(editor.create_memento())

    editor.type("world!")
    history.push(editor.create_memento())

    print("Current content:", editor.get_content())

    editor.delete(6)  # Deleting "world!"
    history.push(editor.create_memento())

    print("Content after deletion:", editor.get_content())

    # Undoing the last change
    last_state = history.pop()
    editor.restore(last_state)
    print("Content after undo:", editor.get_content())

    # Undoing again
    last_state = history.pop()
    editor.restore(last_state)
    print("Content after undo again:", editor.get_content())
```

```python
class Shape:
    def draw(self):
        pass

class Circle(Shape):
    def draw(self):
        print("Circle: draw() method")

class Rectangle(Shape):
    def draw(self):
        print("Rectangle: draw() method")

class Square(Shape):
    def draw(self):
        print("Square: draw() method")

class ShapeFactory:
    @staticmethod
    def create_shape(shape_type):
        if shape_type == "Circle":
            return Circle()
        elif shape_type == "Rectangle":
            return Rectangle()
        elif shape_type == "Square":
            return Square()
        else:
            raise ValueError("Invalid shape type")

if __name__ == "__main__":
    factory = ShapeFactory()

    circle = factory.create_shape("Circle")
    circle.draw()

    rectangle = factory.create_shape("Rectangle")
    rectangle.draw()

    square = factory.create_shape("Square")
    square.draw()
```

```python
from abc import ABC, abstractmethod

# Product interface
class PaymentMethod(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

# Concrete Products
class CreditCard(PaymentMethod):
    def pay(self, amount):
        print(f"Paying {amount} via Credit Card")

class PayPal(PaymentMethod):
    def pay(self, amount):
        print(f"Paying {amount} via PayPal")

class Bitcoin(PaymentMethod):
    def pay(self, amount):
        print(f"Paying {amount} via Bitcoin")

# Factory
class PaymentMethodFactory:
    @staticmethod
    def create_payment_method(method):
        if method == "CreditCard":
            return CreditCard()
        elif method == "PayPal":
            return PayPal()
        elif method == "Bitcoin":
            return Bitcoin()
        else:
            raise ValueError("Invalid payment method")

# Client code
if __name__ == "__main__":
    factory = PaymentMethodFactory()

    # Create payment methods
    payment1 = factory.create_payment_method("CreditCard")
    payment2 = factory.create_payment_method("PayPal")
    payment3 = factory.create_payment_method("Bitcoin")

    # Use payment methods
    payment1.pay(100)
    payment2.pay(50)
    payment3.pay(200)
```

#### builder 
* used to build complex objects 
* separate the build of object from representation
* can use same builder to build different representations.
* builder as factory but if it complex use builder

```python
class b:
    def __init__(self, obj):
        self.builder = obj

    def build_a(self):
        self.builder.parts.append('a')
        return self

    def build_b(self):
        self.builder.parts.append('b')
        return self

    def get(self):
        return self.builder

class builder:
    def __init__(self):
        self.parts = []

    def build(self):
        return b(self).build_a().build_b().get()

b_obj = builder()
result = b_obj.build()
print(result.parts)  # Output: ['a', 'b']
```

```python
class Product:
    def __init__(self):
        self.parts = []

    def add_part(self, part):
        self.parts.append(part)

    def list_parts(self):
        print(f"Product parts: {', '.join(self.parts)}")


class Builder:
    def build_part_a(self):
        pass

    def build_part_b(self):
        pass

    def build_part_c(self):
        pass


class ConcreteBuilder(Builder):
    def __init__(self):
        self.product = Product()

    def build_part_a(self):
        self.product.add_part("Part A")

    def build_part_b(self):
        self.product.add_part("Part B")

    def build_part_c(self):
        self.product.add_part("Part C")

    def get_product(self):
        return self.product


class Director:
    def __init__(self, builder):
        self.builder = builder

    def construct(self):
        self.builder.build_part_a()
        self.builder.build_part_b()
        self.builder.build_part_c()


if __name__ == "__main__":
    builder = ConcreteBuilder()
    director = Director(builder)

    director.construct()

    product = builder.get_product()
    product.list_parts()
```

```python
class Computer:
    def __init__(self):
        self.case = None
        self.motherboard = None
        self.cpu = None
        self.memory = None
        self.storage = None
        self.graphics_card = None

    def display(self):
        print("Computer Specifications:")
        print(f"Case: {self.case}")
        print(f"Motherboard: {self.motherboard}")
        print(f"CPU: {self.cpu}")
        print(f"Memory: {self.memory}")
        print(f"Storage: {self.storage}")
        print(f"Graphics Card: {self.graphics_card}")


class ComputerBuilder:
    def __init__(self):
        self.computer = Computer()

    def set_case(self, case):
        self.computer.case = case

    def set_motherboard(self, motherboard):
        self.computer.motherboard = motherboard

    def set_cpu(self, cpu):
        self.computer.cpu = cpu

    def set_memory(self, memory):
        self.computer.memory = memory

    def set_storage(self, storage):
        self.computer.storage = storage

    def set_graphics_card(self, graphics_card):
        self.computer.graphics_card = graphics_card

    def get_computer(self):
        return self.computer


class Director:
    def __init__(self, builder):
        self.builder = builder

    def build_basic_computer(self):
        self.builder.set_case("Basic Case")
        self.builder.set_motherboard("Basic Motherboard")
        self.builder.set_cpu("Basic CPU")
        self.builder.set_memory("Basic Memory")
        self.builder.set_storage("Basic Storage")
        self.builder.set_graphics_card("Basic Graphics Card")

    def build_custom_computer(self):
        self.builder.set_case("Custom Case")
        self.builder.set_motherboard("Custom Motherboard")
        self.builder.set_cpu("Custom CPU")
        self.builder.set_memory("Custom Memory")
        self.builder.set_storage("Custom Storage")
        self.builder.set_graphics_card("Custom Graphics Card")


if __name__ == "__main__":
    builder = ComputerBuilder()
    director = Director(builder)

    print("Building Basic Computer:")
    director.build_basic_computer()
    basic_computer = builder.get_computer()
    basic_computer.display()

    print("\nBuilding Custom Computer:")
    director.build_custom_computer()
    custom_computer = builder.get_computer()
    custom_computer.display()
```

#### single tone 
* class only have one instance 
* provide global access to instance
* should not contain any references to self but use static 

```python
class Singleton:
    _instance = None

    def __new__(cls, *args, **kwargs):
        if not cls._instance:
            cls._instance = super().__new__(cls, *args, **kwargs)
        return cls._instance


if __name__ == "__main__":
    # Creating instances
    singleton1 = Singleton()
    singleton2 = Singleton()

    print("ID of singleton1:", id(singleton1))
    print("ID of singleton2:", id(singleton2))
    print("Are they the same instance?", singleton1 is singleton2)
```

```python
class Log:
    _instance = None

    def __new__(cls):
        if cls._instance is None: ### if object empty use cls 
            cls._instance = super().__new__(cls) 
        return cls._instance

    def print_details(self, message):
        print(f"Log: {message}")

l1 = Log()
l2 = Log()

print(l1 is l2)  # Output: True

l1.print_details("m1")
l2.print_details("m2")
```
